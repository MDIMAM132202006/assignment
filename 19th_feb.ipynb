Q1. What is Flask Framework? What are the advantages of Flask Framework?

ans: 

**Flask** is like a toolbox for building websites and web applications using the Python programming language. It's known for being simple and flexible.

**Advantages of Flask:**

1. **Easy to Use:** It's like a beginner-friendly tool for creating websites.

2. **You Choose:** You can pick and choose the tools you need for your project. It's like building with LEGO blocks - you decide which pieces to use.

3. **Add More Features:** You can add extra features to your website, like user logins or a database, with the help of special add-on tools.

4. **Good for Small to Medium Projects:** Flask is great for making smaller websites or apps, not just massive ones.

5. **No Complicated Rules:** There are not too many strict rules to follow, so you have more freedom to build your site the way you want.

6. **Easy to Learn:** It's like learning to ride a bike with training wheels. It's a great way to start web development.

7. **Lots of Help:** Many people use Flask, so you can easily find guides, tutorials, and support online.

8. **Makes APIs:** If you want to create a system for apps to talk to each other (like a food delivery app talking to a restaurant app), Flask can do that too.

9. **Test Your Work:** Flask helps you check that your website works correctly, like making sure all the links and buttons do what they should.

10. **Free and Open:** It's completely free to use, and you can even help make it better if you want!

Think of Flask as a friendly and straightforward tool that helps you build web applications, and it's particularly great for getting started with web development.

Q3. What is App routing in Flask? Why do we use app routes?


ans:
    App routing in Flask is the process of defining how different URLs should be handled by your web application. It allows you to specify which Python functions should be executed when specific URLs are requested. App routes are used to:

1. Define URL handling and request processing.
2. Encourage a modular and organized application design.
3. Handle dynamic URLs and parameters.
4. Create RESTful APIs with clear endpoints.
5. Customize error handling for error pages.

App routes are defined using the `@app.route()` decorator, and they play a crucial role in structuring and controlling the behavior of your Flask web application.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

ans:
    
In Flask, the url_for() function is used for URL building.

from flask import Flask, url_for

app = Flask(__name)

@app.route('/')

def home():
    return 'Welcome to the home page.'

@app.route('/about')

def about():
    return 'This is the about page.'

@app.route('/contact')

def contact():
    return 'Contact us at example@email.com'

if __name__ == '__main__':
    with app.test_request_context():
    
        # Using url_for to generate URLs
        home_url = url_for('home')
        about_url = url_for('about')
        contact_url = url_for('contact')
        
        # Printing the generated URLs
        print(f'Home URL: {home_url}')
        print(f'About URL: {about_url}')
        print(f'Contact URL: {contact_url}')
